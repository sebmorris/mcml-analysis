```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt

data = loadmat('./data/29-Nov-2022 120415 Seb reference Patient')
refs = data['Spectra']
wavelengths = data['Wavelengths']

ref = refs[0, :]
ref = ref / np.max(ref)

plt.figure(figsize=(4,4))
plt.plot(range(1024), ref)
plt.show()
```

```python
refs = loadmat('./data/29-Nov-2022 115640 Seb 3cm Patient')['Spectra']

ref_3_cm = refs[0, :]
ref_3_cm = ref_3_cm / np.max(ref_3_cm)

plt.figure(figsize=(4,4))
plt.plot(range(1024), ref_3_cm)
plt.show()
```

```python
refs = loadmat('./data/29-Nov-2022 115434 Seb 1cm Patient')['Spectra']

ref_1_cm = refs[0, :]
ref_1_cm = ref_1_cm / np.max(ref_1_cm)

plt.figure(figsize=(4,4))
plt.plot(range(1024), ref_1_cm)
plt.show()
```

```python
np.save('./data/ref', ref)
np.save('./data/1cm', ref_1_cm)
np.save('./data/3cm', ref_3_cm)

np.save('./data/wavelengths', wavelengths)
```
