```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 100)
y = 3*x + 2 + np.random.normal(loc=0, scale=0.1, size=len(x))

plt.figure(figsize=(6,6))
plt.scatter(x, y)
plt.xlabel('x / arb. units')
plt.ylabel('y / arb. units')
plt.text(0.05, 4.5, r'$y \sim \mathcal{N}(ax + b, 0.1)$', fontsize=14)
```
